```python
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
from experiment.cliffs_delta import cliffs_delta
```

```python
native = re.compile('.*?hui\.android.*?')
capacitor = re.compile('.*?hui\.capacitor.*?')
scrolling1 = re.compile('.*?hui\.scrolling1')
scrolling4 = re.compile('.*?hui\.scrolling4')
scrolling10 = re.compile('.*?hui\.scrolling10')
```

```python
column_names = [
    'app',
    'device_model',
    'interaction',
    'idle1_energy',
    'idle1_duration',
    'treatment_energy',
    'treatment_duration',
    'idle2_energy',
    'idle2_duration'
]
df = pd.read_csv('../data/measurements.txt', sep=';', header=None, names=column_names)
```

## Data preprocessing

```python
def real_energy(row):
    
    # real energy
    idle1 = row['idle1_energy'] / row['idle1_duration']
    idle2 = row['idle2_energy'] / row['idle2_duration']
    idle_mean = (idle1 + idle2) / 2
    row['idle_mean'] = idle_mean
    row['energy'] = row['treatment_energy'] - (row['treatment_duration'] * idle_mean)
    
    # app interaction
    app_interaction = 'hybrid-ui-' + row['interaction']
    app_short = 'Hybrid UI'
    
    if native.match(row['app']):
        app_interaction = 'native-' + row['interaction']
        app_short = 'Android native'
    elif capacitor.match(row['app']):
        app_interaction = 'capacitor-' + row['interaction']
        app_short = 'Capacitor'
    elif scrolling10.match(row['app']):
        app_interaction = 'hybrid-ui-scrolling-10'
        app_short = 'Hybrid UI (10)'
    elif scrolling4.match(row['app']):
        app_interaction = 'hybrid-ui-scrolling-4'
        app_short = 'Hybrid UI (4)'
    elif scrolling1.match(row['app']):
        app_interaction = 'hybrid-ui-scrolling-1'
        app_short = 'Hybrid UI (1)'
    
    row['app_interaction'] = app_interaction
    row['app_short'] = app_short
    
    return row

processed_df = df.apply(real_energy, axis=1)
```

## Mean differences

```python
energy_means = processed_df[['app_interaction', 'energy']].groupby(['app_interaction']).mean()
energy_means
```

```python
def calc_energy_diff(series, item1, item2):
    diff = float((energy_means.loc[item2] - energy_means.loc[item1]) / energy_means.loc[item1])
    
    if diff < 0:
        print("{} uses {:.1f}% less energy than {}".format(item2, diff * -100, item1))
    else:
        print("{} uses {:.1f}% more energy than {}".format(item2, diff * 100, item1))

calc_energy_diff(energy_means, 'capacitor-dialog', 'hybrid-ui-dialog')
calc_energy_diff(energy_means, 'capacitor-drawer', 'hybrid-ui-drawer')
calc_energy_diff(energy_means, 'capacitor-scrolling', 'hybrid-ui-scrolling-1')
calc_energy_diff(energy_means, 'capacitor-scrolling', 'hybrid-ui-scrolling-4')
calc_energy_diff(energy_means, 'capacitor-scrolling', 'hybrid-ui-scrolling-10')
calc_energy_diff(energy_means, 'capacitor-sheet', 'hybrid-ui-sheet')
```

## Shapiro-Wilk test for normality
`p < 0.05` shows evidence for non-normal distribution

```python
app_interactions = processed_df['app_interaction'].unique()
non_normality_count = 0

for app_interaction in app_interactions:
    result = stats.shapiro(processed_df[processed_df['app_interaction'] == app_interaction]['energy'])
    print("{} p-value: {:.5f}".format(app_interaction, result[1]))
    
    if result[1] < 0.05:
        non_normality_count += 1
    
print("Non normal samples: {}, normal samples: {}".format(non_normality_count, len(app_interactions) - non_normality_count))    
```

## Two-sided permutation test 

```python
tests = {
    'dialog': ['hybrid-ui-dialog', 'native-dialog', 'capacitor-dialog'],
    'drawer': ['hybrid-ui-drawer', 'native-drawer', 'capacitor-drawer'],
    'scrolling': ['hybrid-ui-scrolling-10', 'native-scrolling', 'capacitor-scrolling'],
    'sheet': ['hybrid-ui-sheet', 'native-sheet', 'capacitor-sheet']
}

def statistic(x, y, axis):
    return np.mean(x, axis=axis) - np.mean(y, axis=axis)

for interaction in tests.keys():   
    hui = processed_df[processed_df['app_interaction'] == tests[interaction][0]]['energy']
    native = processed_df[processed_df['app_interaction'] == tests[interaction][1]]['energy']
    capacitor = processed_df[processed_df['app_interaction'] == tests[interaction][2]]['energy']
    
    native_r = stats.permutation_test((
        hui,
        native
    ), statistic=statistic, vectorized=True, n_resamples=100000)
    
    native_c = cliffs_delta(hui, native)
    
    capacitor_r = stats.permutation_test((
        hui,
        capacitor
    ), statistic=statistic, vectorized=True, n_resamples=100000)
    
    capacitor_c = cliffs_delta(hui, capacitor)
    
    print(interaction)
    print("  native: p-value={:.5f}, eff={:.5f}".format(native_r.pvalue, native_c[0]))
    print("  capacitor: p-value={:.5f}, eff={:.5f}".format(capacitor_r.pvalue, capacitor_c[0]))
```

## Boxplot energy consumption 

```python
interactions = ['dialog', 'drawer', 'scrolling', 'sheet']

fig, axs = plt.subplots(1, 4, sharey=True, figsize=(18, 8))

for i, interaction in enumerate(interactions):    
    processed_df[processed_df['interaction'] == interaction][['app_short', 'energy', 'interaction']].boxplot(
        ax=axs[i],
        column=['energy'],
        by=['app_short'],
        widths=.5,
        positions=[0, 1, 2, 4, 3] if i == 2 else [0, 1, 2],
        showfliers=False)
    
    axs[i].set_title(interaction.capitalize(), fontsize='xx-large', pad=15)
    axs[i].set_xlabel('')
    
    for label in axs[i].get_xticklabels():
        label.set_rotation(60)
        label.set_ha('right')
        label.set_fontsize('xx-large')

for label in axs[0].get_yticklabels():
    label.set_fontsize('x-large')  

axs[0].set_ylabel("Joules", rotation=90, labelpad=7, fontsize='xx-large')
axs[0].set_ylim(bottom=0)
        
fig.suptitle('')
fig.tight_layout()
fig.subplots_adjust(wspace=0.1, hspace=0.1)
fig.savefig('../docs/results.png', dpi=800)
```
